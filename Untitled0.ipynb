Este programa te permitirá convertir entre diferentes monedas. He incluido algunas tasas de cambio comunes, pero ten en cuenta que estas tasas son fijas y no se actualizan en tiempo real. Para obtener tasas de cambio en tiempo real, se necesitaría integrar una API de terceros.

```python
import math

def get_exchange_rate(source_currency, target_currency, rates):
    """Obtiene la tasa de cambio entre dos monedas."""
    source = source_currency.upper()
    target = target_currency.upper()

    if source == target:
        return 1.0

    if source in rates and target in rates[source]:
        return rates[source][target]
    elif target in rates and source in rates[target]:
        # If direct rate not found, try inverse
        return 1 / rates[target][source]
    else:
        return None

def convert_currency():
    """Función principal para la conversión de moneda."""
    # Tasas de cambio (estas son tasas fijas y aproximadas para demostración)
    # Un escenario real requeriría una API para tasas en tiempo real.
    exchange_rates = {
        'USD': {'EUR': 0.92, 'COP': 3900.0, 'GBP': 0.79},
        'EUR': {'USD': 1.08, 'COP': 4200.0, 'GBP': 0.86},
        'COP': {'USD': 1/3900.0, 'EUR': 1/4200.0, 'GBP': 1/4900.0},
        'GBP': {'USD': 1.27, 'EUR': 1.16, 'COP': 4900.0}
    }

    print("\n--- Conversor de Monedas ---")
    print("Monedas disponibles: USD, EUR, COP, GBP")

    while True:
        source_currency = input("Ingrese la moneda de origen (ej. USD): ").upper()
        if source_currency in exchange_rates:
            break
        else:
            print(f"Moneda de origen '{source_currency}' no reconocida. Por favor, elija entre {', '.join(exchange_rates.keys())}.")

    while True:
        target_currency = input("Ingrese la moneda de destino (ej. EUR): ").upper()
        if target_currency in exchange_rates:
            break
        else:
            print(f"Moneda de destino '{target_currency}' no reconocida. Por favor, elija entre {', '.join(exchange_rates.keys())}.")

    while True:
        try:
            amount_str = input("Ingrese la cantidad a convertir: ")
            amount = float(amount_str)
            if amount <= 0:
                print("La cantidad debe ser un número positivo.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Por favor, ingrese un número para la cantidad.")

    rate = get_exchange_rate(source_currency, target_currency, exchange_rates)

    if rate is not None:
        converted_amount = amount * rate
        print(f"\n{amount:.2f} {source_currency} es igual a {converted_amount:.2f} {target_currency}")
    else:
        print(f"No se pudo encontrar una tasa de cambio entre {source_currency} y {target_currency}.")

# Llama a la función para ejecutar el programa
convert_currency()
```

### Cómo funciona el código:

1.  **`exchange_rates`**: Este diccionario almacena las tasas de cambio. Las claves de nivel superior son las monedas de origen, y los diccionarios anidados contienen las monedas de destino y sus respectivas tasas. Por ejemplo, `exchange_rates['USD']['EUR']` te daría cuántos EUR son 1 USD.
2.  **`get_exchange_rate(source_currency, target_currency, rates)`**: Esta función auxiliar busca la tasa de cambio directa o inversa entre las dos monedas especificadas. Si las monedas son las mismas, devuelve `1.0`. Si no encuentra una tasa, devuelve `None`.
3.  **`convert_currency()`**: Esta es la función principal que orquesta el programa:
    *   **Entrada del usuario**: Pide al usuario la cantidad a convertir, la moneda de origen y la moneda de destino.
    *   **Validación de entrada**: Incluye bucles `while` para asegurar que el usuario ingrese una cantidad numérica válida y monedas reconocidas.
    *   **Cálculo**: Utiliza la función `get_exchange_rate` para obtener la tasa y luego multiplica la cantidad por esta tasa.
    *   **Salida**: Muestra el resultado de la conversión.
